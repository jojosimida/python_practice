```python
import pandas as pd
url = 'http://www.cwb.gov.tw/V7/observe/rainfall/ha_100.htm'
df = pd.read_html(url)[2]
df
```

```python
df.columns = df.loc[0] #row 0 to column name 
df
```

```python
df = df.drop(0).reset_index(drop=True) #drop row 0 and reset_index
df
```

```python
df['雨量(毫米)'] = df['雨量(毫米)'].astype('float') #string type to float type
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

plt.rcParams['font.family']='DFKai-SB' #顯示中文 for Windows10
df[:15].plot(kind='bar',y='雨量(毫米)',x='測站名稱',figsize=(12,6),fontsize=18,title='雨量top15');
```
